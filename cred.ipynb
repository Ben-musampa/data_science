### Create a fake Data for training and learning Datascience

```python
# import essential library
import numpy as np # numerical python 
import pandas as pd # for data manipulation
import names # generate fake user names (need to install)
```

```python
# create random dataframe whit pandas, numpy and names
etudiant = {"ID":[i+1 for i in range(200)], # for 200 fakes users
            "AGE":list(np.random.randint(low=18, high=80, size=200)),
            "WEIGHT":list(np.random.uniform(low=60, high=120, size=200)),
            "NAMES": [names.get_full_name() for i in range(200)],
            "HEIGHT": list(np.random.uniform(low=1.5, high=2.5, size=200))}
```

```python
data = pd.DataFrame(etudiant)
# read data in data frame
data
```

```python
# we need to add a SENIORITE column to the dataframe, first i create a fuction can help for that

def get_seniorite(age):
    if age>=50:
        return "Old"
    elif age<50 and age>=18:
        return "Adult"
    else:
        return "Minor"
```

```python
# add a colunm to data
data["SENIORITE"]= data['AGE'].apply(get_seniorite)
# display data
data
```

```python
# calculate BMI function (body mass indexes)

def get_bmi(weight, height):
     return weight/(height**2)
```

```python
# create the interprate function to interprete the calcu of previous function

def interprate(row):
    weight = row["WEIGHT"]
    height = row["HEIGHT"]
    
    im = get_bmi(weight, height)
    
    if im <= 18.5:
        return "underweight"
    elif im>=18.5 and  im<=25:
        return "healthy"
    elif im>=25 and im<=30:
        return "overweiht"
    elif im>=30 and im<=35:
        return "obese"
    elif im>=35 and im<=40:
        return "extremely obese"
    else:
        return "extremely obese"
```

```python
# create BMI column in the dataframe

data['BMI']= data.apply(interprate, axis=1)

# read a dataframe
data
```

```python
# count how many different values in df
data['BMI'].value_counts()
```

```python
# save in csv file
data.to_csv("populations_data.csv")
```
