Voici une explication simple et concise des types de graphiques les plus courants, leurs utilisations, et les données appropriées pour chacun :

## 1. Graphique en Lignes
### Données adaptées :
	•	Séries temporelles (exemple : évolution d’une variable dans le temps).
	•	Comparaison de tendances entre plusieurs catégories.
### Quand l’utiliser ?
	•	Suivre l’évolution d’une donnée continue sur une période donnée (ex. chiffre d’affaires mensuel).
	•	Visualiser des tendances ou des fluctuations.
### Exemple :
	•	Variation de la température par jour.
	•	Suivi des ventes hebdomadaires d’un produit.

```python
# importer les librairies

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# création des données

months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
sales = np.random.randint(100, 500, size=12)

# dataframe
data = pd.DataFrame({'Month': months, 'Sales': sales})
data
```

```python
# graphique en ligne
plt.plot(data['Month'], data['Sales'], marker='o')
plt.title('monthly sales')
plt.xlabel('months')
plt.ylabel('sales')
plt.grid()
plt.show()
```

## 2. Graphique en Barres

### Données adaptées :
	•	Données catégoriques (exemple : regroupées par groupes ou catégories).
	•	Comparaison de plusieurs valeurs.

### Quand l’utiliser ?
	•	Comparer des valeurs entre différentes catégories ou groupes (ex. chiffre d’affaires par région).
	•	Identifier les différences majeures entre des entités.

### Exemple :
	•	Population par pays.
	•	Revenus annuels par secteur d’activité.

```python
# population par région 
regions = ['Kin', 'Goma', 'Boma', 'Moanda']
population = np.random.randint(10000, 500000, size=4)

# dataframe
data_bar = pd.DataFrame({'Region': regions, 'Population': population})
data_bar
```

```python
# graphique en barres
plt.bar(data_bar['Region'], data_bar['Population'], color='skyblue')
plt.title('Population par region ')
plt.xlabel('region')
plt.ylabel('population')
plt.show()
```

## 3. Histogramme

### Données adaptées :
	•	Données quantitatives continues.
	•	Distribution d’une variable (découpée en intervalles).

### Quand l’utiliser ?
	•	Visualiser la distribution d’une variable.
	•	Identifier des schémas comme la symétrie, les outliers ou les modes.

### Exemple :
	•	Répartition des âges des clients.
	•	Distribution des revenus d’une population.

```python
# pour la répation de l'age
ages = np.random.randint(18, 62, size=200)

# Histogramme
plt.hist(ages, bins=10, color='skyblue', edgecolor='black')
plt.title('Répartition des âges')
plt.xlabel('âges')
plt.ylabel('fréquence')
plt.show()
```

## 4. Nuage de Points (Scatter Plot)

### Données adaptées :
	•	Deux variables quantitatives (relation entre X et Y).
	•	Données permettant de visualiser des corrélations ou des patterns.

### Quand l’utiliser ?
	•	Montrer la relation entre deux variables.
	•	Identifier des tendances, clusters ou anomalies.

### Exemple :
	•	Relation entre les dépenses publicitaires et les ventes.

```python
ad_spend = np.random.randint(1000, 10000, size=50) # Dépenses publicitaires
sales = ad_spend * 0.1 * np.random.randint(0, 500, size=50) # Ventes avec bruit

plt.scatter(ad_spend, sales, color='green')
plt.title('Relation entre les dépenses publicitaires et les ventes.')
plt.xlabel('Dépenses')
plt.ylabel('Ventes')
plt.show()
```

## 5. Diagramme en Boîte (Boxplot)

### Données adaptées :
	•	Une variable quantitative (données continues) par catégories optionnelles.

### Quand l’utiliser ?
	•	Résumer la distribution d’une variable à l’aide de ses quartiles.
	•	Identifier les outliers et comprendre la dispersion des données.

### Exemple :
	•	Analyse des salaires par département.
	•	Distribution des notes des étudiants par classe.

```python
# distribution des notes des étudiants par classes
classe = ['L1_info', 'L1_stat', 'L1_math', 'L1_physique']
notes = {cl: np.random.randint(12, 20, size=25) for cl in classe}

# dataframe
data_box = pd.DataFrame(notes)
data_box
```

```python
# Diagramme en boite
data_box.boxplot()
```

## 6. Diagramme Circulaire (Pie Chart)

### Données adaptées :
	•	Données catégoriques représentant des proportions ou des pourcentages.

### Quand l’utiliser ?
	•	Visualiser la répartition proportionnelle d’une catégorie par rapport à un tout.
	•	Cas où le nombre de catégories est limité (idéalement moins de 6).

### Exemple :
	•	Répartition des parts de marché.
	•	Proportion des dépenses dans un budget.

```python
# répartion des parts de marché

compagnies = ['A', 'B', 'C', 'D']
parts_mache = [25, 30, 20, 25]

# Diagramme circulaire
plt.pie(parts_mache, labels=compagnies, autopct='%1.1f%%', startangle=140, colors=['red', 'blue', 'green', 'orange'])
plt.title('parts de marché')
plt.show()
```

## 7. Carte de Chaleur (Heatmap)

### Données adaptées :
	•	Matrices ou données catégoriques croisées (exemple : tableau de corrélation).
	•	Variables montrant des intensités ou des fréquences.

### Quand l’utiliser ?
	•	Identifier des motifs ou des relations d’intensité.
	•	Visualiser des corrélations.

### Exemple :
	•	Tableau de corrélation entre variables d’un dataset.
	•	Fréquences de visites d’un site web par heure et jour.

```python
# Pour calcluler la corrélation entre les variables nous utiliserons la librairies seaborn
import seaborn as sns
```

```python
# création des données
dt = np.random.rand(10, 5)
columns = ['Val_1', 'Val_2', 'Val_3', 'Val_4', 'Val_5']

# dataframe
data_corr = pd.DataFrame(dt, columns=columns)

# la matrice de corrélation
corr = data_corr.corr()
corr
```

```python
# Heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

## 8. Graphe à Barres Empilées

### Données adaptées :
	•	Données catégoriques réparties en sous-catégories.

### Quand l’utiliser ?
	•	Comparer des valeurs globales, tout en visualisant la contribution de chaque sous-catégorie.

### Exemple :
	•	Répartition des ventes par produit et par trimestre.
	•	Proportions de genres dans différentes tranches d’âge.


```python
# ventes par produit et par trimestre
produits = ['Produit 1', 'Produit 2', 'Produit 3']
q1 = np.random.randint(100, 300, size=3)
q2 = np.random.randint(150, 400, size=3)
q3 = np.random.randint(200, 500, size=3)

# dataframe
data_stacked=pd.DataFrame({'Produits': produits, 'Q1': q1, 'Q2': q2, 'Q3': q3}).set_index('Produits')
data_stacked
```

```python
# Barres empilées
data_stacked.plot(kind="bar", stacked=True, color=['blue', 'orange', 'green'])
plt.title('ventes par produit et par trimestre')
plt.ylabel('Ventes')
plt.show()
```

# Conseil général :

## Le choix du graphique dépend de :
	•	La nature des données (quantitatives ou qualitatives).
	•	La question à répondre (relation, distribution, tendance, comparaison, etc.).
	•	La simplicité : Choisissez toujours un graphique facile à interpréter pour votre public.
